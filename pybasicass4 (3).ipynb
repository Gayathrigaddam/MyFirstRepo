```python
#program to find lcm
def lcm(*args):
    maxVal=1
    for i in args:
        maxVal*=i
    lcm=maxVal
    for i in range(max(args),maxVal):
        for j in range(len(args)):
            if(i%args[j]==0):
                continue 
            else:
                break
        else:
            lcm=i
            break
    return lcm
```

```python
#program to find hcf
def hcf(*args):
    hcf=1
    for i in range(2,max(args)):
        for j in range(len(args)):
            if(args[j]%i==0):
                continue 
            else:
                break
        else:
            hcf=i
    return hcf
```

```python
#program to given binary,hexadecimal,octal values of a decimal number
def ConvertToBinary(num):
    return bin(num)[2:]
def ConvertToHexadecimal(num):
    return hex(num)[2:]
def ConvertToOctal(num):
    return oct(num)[2:]
```

```python
num=int(input("Enter the number"))
print("Binary value of {0} is {1}".format(num,ConvertToBinary(num)))
print("Octal value of {0} is {1}".format(num,ConvertToOctal(num)))
print("Hexadecimal value of {0} is {1}".format(num,ConvertToHexadecimal(num)))
```
