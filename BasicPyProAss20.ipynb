```python
#BasicAss20
#Question1
def filter_list(lst):
    lstout=[]
    for i in lst:
        if(type(i)==int):
            lstout.append(i)
    return lstout
```

```python
print(filter_list([1,2,3,'a','b',4]))
print(filter_list(['a',0,1729,'python','1729']))
print(filter_list(['nothing','here']))
```

```python
#Question2
def add_indexes(lst):
    return [i+lst[i] for i in range(len(lst))]
print(add_indexes([0,0,0,0,0]))
print(add_indexes([1,2,3,4,5]))
print(add_indexes([5,4,3,2,1]))
```

```python
#Question3
import math
def cone_volume(height,radius):
    return round(math.pi*radius*radius*height/3 , 2)
print(cone_volume(3,2))
print(cone_volume(15,6))
print(cone_volume(18,0))
```

```python
#Question4
def triangle(num):
    return int(num*(num+1)/2)
print(triangle(1))
print(triangle(6))
print(triangle(215))
```

```python
#Question5
def missing_num(lst):
    for i in range(1,11):
        if i not in lst:
            return i
print(missing_num([1,2,3,4,6,7,8,9,10]))
print(missing_num([1,2,3,4,5,6,7,8,9]))
print(missing_num([1,2,3,4,5,6,8,9,10]))
```
