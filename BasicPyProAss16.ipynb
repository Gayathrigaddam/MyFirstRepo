```python
#BasicAss16
#Question1
def shutter(strin):
    return strin[0:2]+"... "+strin[0:2]+"... "+strin+"?"
```

```python
print(shutter("incredible"))
print(shutter("enthusiastic"))
print(shutter("outstanding"))
```

```python
#Question2
import math
def radians_to_degerees(rad):
    return round(180*rad/math.pi,1)
```

```python
print(radians_to_degerees(1))
print(radians_to_degerees(20))
print(radians_to_degerees(50))
```

```python
#Question3
def is_curzon(num):
    return ((1+(2**num))%(1+(2*num))==0)
```

```python
print(is_curzon(5))
print(is_curzon(10))
print(is_curzon(14))
```

```python
#Question4
def area_of_hexagon(s):
    return round((3**1.5)*s*s/2,1)
```

```python
print(area_of_hexagon(1))
print(area_of_hexagon(2))
print(area_of_hexagon(3))
```

```python
#Question5
def binary(num):
    return bin(num)[2:]
```

```python
print(binary(1))
print(binary(5))
print(binary(10))
```
