```python
#6
from collections import OrderedDict
fancy=OrderedDict({'a':1,'b':2,'d':4,'c':3})
print(fancy)
```

```python
#Yes printed in the same order in this case.But OrderedDict preserves the order in which the items are inserted and prints them 
#in order,whereas a plain dictionary doesn't preserve the order and prints them randomly
```
