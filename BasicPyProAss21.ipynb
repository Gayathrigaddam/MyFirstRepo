```python
#BasicAss21
#Question1
def next_in_line(lst,num):
    if(len(lst)==0):
        return "No list has been selected"
    else:
        lst=lst[1:]
        lst.append(num)
        return lst
print(next_in_line([5,6,7,8,9],1))
print(next_in_line([7,6,3,27,17],10))
print(next_in_line([1,10,20,42],6))
print(next_in_line([],1))
```

```python
#Question2
def get_budgets(lstdictin):
    sum=0
    for i in lstdictin:
        sum+=i['budget']
    return sum
print(get_budgets([{"name":"John","age":21,"budget":23000},{"name":"Steve","age":32,"budget":40000},{"name":"Martin","age":16,"budget":2700}]))
print(get_budgets([{"name":"John","age":21,"budget":29000},{"name":"Steve","age":32,"budget":32000},{"name":"Martin","age":16,"budget":1600}]))
```

```python
#Question3
def alphabet_soup(strin):
    return "".join(sorted(list(strin)))
print(alphabet_soup("hello"))
print(alphabet_soup("edabit"))
print(alphabet_soup("hacker"))
print(alphabet_soup("geek"))
print(alphabet_soup("javascript"))
```

```python
#Question4
def compound_interest(prin,years,rate,times):
    sum=prin
    rate=rate/times
    for i in range(int((years)*times)):
        sum=prin*(1+rate)
        prin=sum
    return round(sum,2)
```

```python
print(compound_interest(10000,10,0.06,12))
print(compound_interest(100,1,0.05,1))
print(compound_interest(3500,15,0.1,4))
print(compound_interest(100000,20,0.15,365))
```

```python
#Question5
def isnum(num):
    if(type(num)==int):
        return True
    else:
        return False
def return_only_integer(lst):
    return filter(isnum,lst)

lst=[]
for s in return_only_integer([9,2,"space","car","lion",16]):
    lst.append(s)
print(lst)
lst=[]
for s in return_only_integer([81,"hello","basketball","fox",16]):
    lst.append(s)
print(lst)
lst=[]
for s in return_only_integer([10,'121',56,20,"car","lion",13]):
    lst.append(s)
print(lst)
  
```
