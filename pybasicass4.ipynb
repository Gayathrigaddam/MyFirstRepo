```python
#program to find factorial of a number
num=int(input("Enter the number"))
fact=1
if(num<0):
    print("Factorial doesn't exist for a negative number")
elif(num==0):
    print("Factorial is 1")
else:
    while(num!=1):
         fact*=num
         num-=1
    print("Factorial of the given number : ",fact)
```

```python
#program to display the multiplication table
num=int(input("Enter the number"))
count=int(input("Enter till which number you want the table"))
for i in range(1,count+1):
    print("{0} * {1} = {2}".format(num,i,num*i))
```

```python
#write a program to print fibanocci sequence
num=int(input("How many terms you want in the fibanocci sequence :"))
l=[0,1]
print(l[0]+l[1])
for i in range(2,num):
    l.append(l[i-1]+l[i-2])
print(l)
```

```python
#write a program to check armstrong number
num=int(input("Enter the number"))
s=str(num)
length=len(s)
sum=0
for i in range(length):
    sum+=int(s[i])**length
if(sum==num):
    print("Yes!given number is an armstrong number")
else:
    print("Sorry!given number is not an armstrong number")
```

```python
#write a program to find armstrong number in given interval
def checkArmstrong(num):
    s=str(num)
    length=len(s)
    sum=0
    for i in range(length):
            sum+=int(s[i])**length
    if(sum==num):
        return 1
    else:
        return 0
lower=int(input("Enter the lower bound of the interval"))
upper=int(input("Enter the upper bound of the interval"))
l=[]
for i in range(lower,upper+1):
    if(checkArmstrong(i)):
        l.append(i)
print(l)
```
