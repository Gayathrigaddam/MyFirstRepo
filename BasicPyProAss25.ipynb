```python
#Question1
def equal(a,b,c):
    lst=[[a,b,c].count(i) for i in [a,b,c] if [a,b,c].count(i)!=1]
    res=0
    if(len(lst)!=0):
        res=max(lst)
    return res

print(equal(1,1,1))
print(equal(3,4,3))
print(equal(3,4,1))
```

```python
#Question2
def dict_to_list(dictin):
    lst=[]
    for i,j in dictin.items():
        k=(i,j)
        lst.append(k)
    return sorted(lst)
print(dict_to_list({"D":1,"B":2,"C":3}))
print(dict_to_list({"likes":100,"dislikes":2,"followers":320}))
```

```python
#Question3
def mapping(lstin):
    dictout={}
    for i in lstin:
        dictout[i]=i.upper()
    return dictout

print(mapping(['p','s']))
print(mapping(['a','b','c']))
print(mapping(['a','v','y','z']))
```

```python
#Question4
def vow_replace(strin,char):
    lst=list(strin)
    for i in range(len(lst)):
        if(lst[i] in "aeiou"):
            lst[i]=char
    return "".join(lst)
```

```python
print(vow_replace("apple and bananas",'u'))
print(vow_replace("cheese casserole",'o'))
print(vow_replace("stuffed jalapeno poppers",'e'))
```

```python
#Question5
def ascii_capitalize(strin):
    strout=""
    for i in list(strin):
        if(ord(i)%2==0):
            strout+=i.upper()
        else:
            strout+=i.lower()
    return strout
```

```python
print(ascii_capitalize("to be or not to be!"))
print(ascii_capitalize("THE LITTLE MERMAID"))
print(ascii_capitalize("Oh what a beautiful morning."))
```
