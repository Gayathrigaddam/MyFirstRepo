```python
#program to swap 2 variables without temp
a=int(input("Enter a"))
b=int(input("Enter b"))
print("Value of a before swap:",a)
print("Value of b before swap:",b)
a,b=b,a
print("Value of a after swap:",a)
print("Value of b after swap:",b)
```

```python
#program to convert kilometers to miles
kilometers=int(input("Enter distance in kms:"))
miles=kilometers*0.62137
print("Distance in miles = ",miles)
```

```python
#program to convert Celsius to fahrenheit
celcius=int(input("Enter temperature in celcius"))
fahrenheit=(celcius*1.8)+32
print("Temperature in fahrenheit=",fahrenheit)
```

```python
#program to display calendar
import calendar
year=int(input("Enter the year"))
month=int(input("Enter the month"))
print("Calendar for given month year is:")
print(calendar.month(year,month))
```

```python
# python program to solve quadratic equation
a=int(input("Enter co-efficient of x^2: "))
b=int(input("Enter co-efficient of x: "))
c=int(input("Enter constant value: "))
discrmnt=(b*b-(4*a*c))**0.5
sol1,sol2=((-b+discrmnt)/2*a,(-b-discrmnt)/2*a)
print("Solutions of the equation are :",sol1,sol2)
```
