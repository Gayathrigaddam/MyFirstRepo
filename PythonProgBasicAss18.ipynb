```python
#Ques1
def filter_list(lst):
    rem=[]
    for i in lst:
        if(type(i)!=str):
              rem.append(i)
    return rem
print(filter_list([1,2,'a','b']))
print(filter_list([1,'a','b',0,15]))
print(filter_list([1,2,'aasf','1','123','123']))
```

```python
#Ques2
def reverse(inp1):
   out=inp1[::-1]
   res=''
   for i in out:
     if(i.isupper()):
        res+=i.lower()
     else:
        res+=i.upper()
   return res
print(reverse('Hello World'))
print(reverse('ReVeRsE'))
print(reverse('Radar'))
```

```python
#Ques4
def factorial(n):
    if(n==0 or n==1):
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
print(factorial(3))
print(factorial(1))
print(factorial(0))
```

```python
#Ques5
def move_to_end(inp1,inp2):
         l2=[]
         count=0
         for i in inp1:
              if(i==inp2):
                    count+=1
              else:
                l2.append(i)
         inp2=[inp2]   
         return l2+list((inp2*count))   
print(move_to_end([1,3,2,4,4,1],1))
print(move_to_end([7,8,9,1,2,3,4],9))
print(move_to_end(['a','a','a','b'],'a'))
```
