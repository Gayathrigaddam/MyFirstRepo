```python
#BasicProPyAss24
#Question1
def amplify(n):
    lstout=[]
    for i in range(1,n+1):
        if(i%4==0):
            lstout.append(i*10)
        else:
            lstout.append(i)
    return lstout
```

```python
print(amplify(4))
print(amplify(3))
print(amplify(25))
```

```python
#Question2
def unique(lst):
    for i in lst:
        if(lst.count(i)==1):
            return i
```

```python
print(unique([3,3,3,7,3,3]))
print(unique([0,0,0.77,0,0]))
print(unique([0,1,1,1,1,1,1,1]))
```

```python
#Question3
import math
class Circle:
    def __init__(self,rad):
        self.rad=rad
    def getArea(self):
        return round(math.pi*self.rad*self.rad)
    def getPerimeter(self):
        return round(2*math.pi*self.rad)
```

```python
circy=Circle(11)
circy.getArea()
```

```python
circy=Circle(4.44)
circy.getPerimeter()
```

```python
#Question4
def sort_by_length(lst):
    lst=list(sorted(lst,key=len))
    return lst
```

```python
print(sort_by_length(["Google","Apple","Microsoft"]))
print(sort_by_length(["Leonardo","Michaelangelo","Raphael","Donatello"]))
print(sort_by_length(["Jung",'Tiring','Einstein']))
```

```python
#Question5
def is_triplet(a,b,c):
    lst=[a,b,c]
    mx=max(lst)
    lst.remove(mx)
    return((lst[0]**2)+(lst[1]**2)==mx**2)
```

```python
print(is_triplet(3,4,5))
print(is_triplet(13,12,5))
print(is_triplet(3,2,1))
```
