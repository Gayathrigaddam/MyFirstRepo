```python
#BasicAss15
#Question1
def func(n):
    for i in range(n):
        if (i%5==0 and i%7==0):
            yield i
```

```python
str1=""
for i in func(100):
    str1+=str(i)+','
print(str1[:-1])
```

```python
#Question2
def func(n):
    for i in range(n+1):
        if (i%2==0):
            yield i
```

```python
str1=""
for i in func(10):
    str1+=str(i)+','
print(str1[:-1])
```

```python
#Question3
def fib(n):
    lst=[]
    lst.append(0)
    lst.append(1)
    [lst.append(lst[i-1]+lst[i-2]) for i in range(2,n+1)]
    return lst
```

```python
str1=""
a=fib(7)
for i in a:
    str1+=str(i)+','
print(str1[:-1])
```

```python
#Question5
class Shape():
    def area(self):
        return 0
class Square(Shape):
    def __init__(self,side):
        self.side=side
    def area(self):
        return self.side**2
```
