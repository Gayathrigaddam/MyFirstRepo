```python
#Ques1
def double_char(inp1):
      out=''
      for i in inp1:
         out+=i*2
      return out
print(double_char('String'))
print(double_char('Hello World!'))
print(double_char('1234!_'))
```

```python
#Ques2
def reverse(inp1):
   if(type(inp1)==bool):
         return not inp1
   else:
        return 'boolean expected'
print(reverse(True))
print(reverse(False))
print(reverse(0))
print(reverse(None))
```

```python
#Ques3
def num_layers(n):
    return str(0.001*(2**(n-1)))+'m'
print(num_layers(1))
print(num_layers(4))
print(num_layers(21))
```

```python
#Ques4
def index_of_caps(inp1):
      lst=[]
      for i in range(len(inp1)):
           if(inp1[i].isupper()):
                 lst.append(i)
      return lst
print(index_of_caps('eDaBiT'))
print(index_of_caps('eQuINoX'))
print(index_of_caps('determine'))
print(index_of_caps('STRIKE'))
print(index_of_caps('sUn'))
```

```python
#Ques5
def find_even_nums(n):
   lst=[]
   for i in range(1,n+1):
       if(i%2==0):
           lst.append(i)
   return lst
print(find_even_nums(8))
print(find_even_nums(4))
print(find_even_nums(2))
```
