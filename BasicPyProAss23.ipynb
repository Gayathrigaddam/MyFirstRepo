```python
#BasicProAss
#Question1
def is_symmetrical(num):
    return (str(num)==str(num)[::-1])
```

```python
print(is_symmetrical(7227))
print(is_symmetrical(12567))
print(is_symmetrical(4444444))
print(is_symmetrical(9939))
```

```python
#Question2
def multiply_nums(strin):
    res=1
    for i in list(map(int,strin.split(","))):
        res*=i
    return res
```

```python
print(multiply_nums("2,3"))
print(multiply_nums("1,2,3,4,5"))
print(multiply_nums("54,75,0"))
print(multiply_nums("-2,30"))
```

```python
#Question3
def square_digits(num):
    out=""
    for i in str(num):
        out+=str(int(i)**2)
    return out
```

```python
print(square_digits(9119))
print(square_digits(2483))
print(square_digits(3212))
```

```python
#Question4
def setify(lst):
    return sorted(set(lst))
```

```python
print(setify([1,3,3,5,5]))
print(setify([4,4,4,4]))
print(setify([5,7,8,9,10,15]))
```

```python
#Question5
def meanfunc(num):
    num=str(num)
    l=len(num)
    sum=0
    for i in num:
        sum+=int(i)
    return int(sum/l)
```

```python
print(meanfunc(42))
print(meanfunc(12345))
print(meanfunc(666))
```
